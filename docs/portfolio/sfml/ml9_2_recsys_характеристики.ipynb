### <b><span style='color:#686dec'>Задание</span></b>

#### <b><span style='color:#686dec'>Введение</span></b> 

Задача рекомендательной системы заключается в том, чтобы информировать пользователя о товарах, которые могут быть ему интересны в текущий момент.

Кому нужны рекомендательные системы?

- `Клиенту` — предоставляет информацию о товарах.
- `Сервису` — помогает зарабатывать на предоставлении услуг.

Предмет рекомендации может варьироваться и включает:

- Товары (например, Amazon, Ozon)
- Статьи (Arxiv.org)
- Новости (Яндекс.Дзен)
- Изображения (500px)
- Видео (YouTube, Netflix)
- Людей (LinkedIn)
- Музыку (Last.fm, Pandora) и плейлисты.

Цель рекомендации может быть различной, например:

- Покупка
- Информирование
- Обучение
- Заведение контактов.

Контекст рекомендации зависит от действий пользователя в данный момент, таких как просмотр товаров, прослушивание музыки или общение.

Источник рекомендации может включать:

- Аудиторию (например, средний рейтинг ресторана на TripAdvisor)
- Схожих по интересам пользователей
- Экспертное сообщество (для сложных товаров, например, вина).

Степень персонализации рекомендаций может быть разной:

- Неперсональные рекомендации — одинаковые для всех.
- Более продвинутые — основаны на данных текущей сессии.
- Персональные рекомендации — используют всю доступную информацию о клиенте.

Формат рекомендации может быть разнообразным: всплывающее окно, отсортированный список или лента внизу экрана.

Алгоритмы рекомендательных систем разнообразны, но все они сводятся к нескольким основным подходам.


#### <b><span style='color:#686dec'>Ассоциативных правил</span></b> 

Алгоритм построения ассоциативных правил основан на статистическом анализе совместной встречаемости товаров в покупательских корзинах. Ассоциативные правила (ARL) можно формулировать как «Если купили x, то также купили y». 

Анализируется набор транзакций, каждая из которых содержит уникальный набор товаров (itemset). С помощью алгоритмов ARL выявляются правила совпадения товаров в одной транзакции, которые затем сортируются по их значимости.

Для оценки силы ассоциативных правил вводятся несколько метрик: Support, Confidence и Lift.

**Support (поддержка)**

- Это метрика, которая показывает, как часто определённая комбинация товаров встречается в наборе транзакций.
- Она рассчитывается как отношение числа транзакций, содержащих данный набор товаров, к общему числу транзакций

Формально, поддержка для набора товаров  X  определяется следующим образом:

- Support(X) = Количество транзакций, содержащих X / Общее количество транзакци

- Например, если у вас есть 1000 транзакций, и 100 из них содержат товары A и B, то поддержка для набора товаров {A, B} будет:

- Support(A, B) = 100 / 1000 = 0.1

- Это означает, что 10% всех транзакций содержат товары A и B вместе. 

Поддержка помогает определить, насколько часто встречается данный набор товаров в данных и является важным параметром для фильтрации незначительных правил.

В ассоциативных правилах `confidence` (доверие) — это мера, которая показывает, насколько часто правило верно, когда условие соблюдено. Она рассчитывается как отношение количества транзакций, содержащих и условие, и следствие, к количеству транзакций, содержащих только условие.

Формально, если у нас есть правило вида  A → B , то confidence рассчитывается по следующей формуле:

Confidence(A → B) = Support(A ∩ B) / Suppor(A)}


где:

-  Support(A ∩ B)  — это количество транзакций, содержащих как  A , так и  B .

-  Support(A)  — это количество транзакций, содержащих только  A .

Таким образом, confidence показывает вероятность того, что  B  произойдет, если  A  уже произошло. Значение confidence варьируется от 0 до 1 (или от 0% до 100%), где более высокие значения указывают на более сильную связь между  A  и  B .

Алгоритм `Apriori` — это один из самых известных алгоритмов для выявления ассоциативных правил в больших наборах данных. Он был предложен в 1994 году и широко используется в задачах анализа данных, таких как анализ покупательского поведения, рекомендательные системы и другие.

▎Основные идеи алгоритма Apriori:

1. Ассоциативные правила: Алгоритм нацелен на нахождение правил вида "если A, то B", где A и B — это наборы элементов (например, товары). Правила помогают понять, какие элементы часто встречаются вместе.

2. Поддержка (Support): Это мера, которая показывает, как часто элемент или набор элементов встречается в данных. Например, если в 100 транзакциях товар A встречается в 20, то поддержка A составляет 20%.

3. Достоверность (Confidence): Это мера, показывающая, насколько часто правило верно. Например, если правило "если A, то B" имеет достоверность 80%, это означает, что в 80% случаев, когда присутствует A, также присутствует B.

▎Принцип работы алгоритма:

1. Генерация частых наборов: Алгоритм начинает с поиска всех одиночных элементов (1-мерных наборов) и вычисляет их поддержку. Затем он отбирает только те элементы, которые соответствуют заданному порогу поддержки (минимальная поддержка).

2. Итеративное расширение: После нахождения всех частых 1-мерных наборов алгоритм использует их для генерации кандидатов 2-мерных наборов. Поддержка этих кандидатов также вычисляется, и отбираются только те, которые соответствуют минимальной поддержке.

3. Повторение: Этот процесс повторяется для наборов размерности 3, 4 и так далее до тех пор, пока не останется кандидатов, которые соответствуют минимальной поддержке.

4. Генерация правил: После того как все частые наборы найдены, алгоритм генерирует ассоциативные правила из этих наборов и вычисляет их достоверность. Правила, которые соответствуют заданному порогу достоверности, сохраняются.

▎Преимущества и недостатки:

Преимущества:

-  Простота и понятность.
- Эффективен для небольших и средних наборов данных.

Недостатки:

- Высокая вычислительная сложность при работе с большими наборами данных из-за необходимости многократного прохода по данным.
- Неэффективен при наличии большого количества редких элементов.
- Можем рекомендовать только старые товары которые встречалисть в совместных покупках 

Алгоритм Apriori является основой для более сложных и эффективных методов, таких как FP-Growth, который решает некоторые из ограничений Apriori.
